# Qubit And Bloch Sphere
In this notebook , we explore the foundational buliding block of quantum computing  - **Qubits** , their **state representation** , and the **bloch sphere**. there are following points cover in this notbook:
- what a qubit is mathematically and geometrically
- how the bloch sphere represents qubit states
- how quantum gate rotates vector on the Bloch sphere
- what superposition look like visually
- custom qubit creation using angle (θ,φ)
- Bloch visualization of **X,Y,Z,H and phase gate**
- how phase changes rotate states around the  Z-axis

This notebook expands your intuition of qubits through **visual learning**.


## 1.Import Libraries

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np
```

## 2. Visualizing |0> and |1> on the Bloch Sphere 

```python
state_0 = Statevector.from_label('0')
state_1 = Statevector.from_label('1')
plot_bloch_multivector([state_0,state_1],figsize=(3,3))
```

## 3. Applying Quantum Gates and Observing Bloch Sphere
we now apply common quantum gates and observe how they rotate the state on Bloch sphere.
- **H-gate** - Moves |0> to the **equator**.
- **X-gate** - flips the qubit accross the X-axis.
- **Y-gate** - changes only the **phase** (rotation around Z-axis).

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.draw()
```

```python
state_h =Statevector.from_instruction(qc)
plot_bloch_multivector(state_h,figsize=(3,3))
```

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw()
state_x =Statevector.from_instruction(qc)
plot_bloch_multivector(state_x,figsize=(3,3))
```

```python
qc = QuantumCircuit(1)
qc.y(0)
qc.draw()
state_y =Statevector.from_instruction(qc)
plot_bloch_multivector(state_y,figsize=(3,3))
```

```python
qc = QuantumCircuit(1)
qc.z(0)
qc.draw()
state_z =Statevector.from_instruction(qc)
plot_bloch_multivector(state_z,figsize=(3,3))
```

## 4.Custom Qubit: Parameterized State

Any Qubit can be written as :


$|\psi \rangle =\cos \left(\frac{\theta }{2}\right)|0\rangle +e^{i\phi }\sin \left(\frac{\theta }{2}\right)|1\rangle$


let's create a custom qubit with **θ = π/3 , φ = π/4**.

```python
theta =np.pi/3
phi = np.pi/4
alpha = np.cos(theta/2)
beta = np.exp(1j*phi) * np.sin(theta/2)
custom_state= Statevector([alpha,beta])
plot_bloch_multivector(custom_state,figsize=(3,3))
```

## 5. Visualizing Rotations around Each Axis
Let's apply rotations :
- **Rx(π/2)** - rotates around X-axis
- **Ry(π/2)** - rotates around Y-axis
- **Rz(π/2)** - rotates around Z-axis


```python
qc =QuantumCircuit(1)
qc.rx(np.pi/2,0)
state_rx=Statevector.from_instruction(qc)
plot_bloch_multivector(state_rx,figsize=(3,3))
```

```python
#rotation around Y-axis
qc=QuantumCircuit(1)
qc.ry(np.pi/2,0)
state_ry=Statevector.from_instruction(qc)
plot_bloch_multivector(state_ry,figsize=(3,3))
```

```python
#rotation around Z-axis
qc=QuantumCircuit(1)
qc.rz(np.pi/2,0)
state_rz=Statevector.from_instruction(qc)
plot_bloch_multivector(state_rz,figsize=(3,3))
```

## 6.Phase Gate Example
The **Phase gate**  add complex phase to state |1> states:

$$ P(\phi) = \begin{bmatrix} 1 & 0 \\\\ 0 & e^{i\phi} \end{bmatrix} $$


This rotates the Qubit around the **Z-axis**.

```python
qc = QuantumCircuit(1)
qc.h(0) #put state on equator
qc.p(np.pi/3, 0) #phase shift
state_p =Statevector.from_instruction(qc)
plot_bloch_multivector(state_p,figsize=(3,3))
```
