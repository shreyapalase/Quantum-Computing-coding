## Import python libraries
We will used Numpy for linear algebra and complex vector operations.

```python
import numpy as np
np.set_printoptions(precision =3,suppress=True)
```

# 2. Qubit As vectors
A single qubit represented as a 2-D complex vectors.
$$ |0> = \begin{pmatrix} 1 \\\\ 0 \end{pmatrix}, \quad |1> = \begin{pmatrix} 0 \\\\ 1 \end{pmatrix} $$

```python
ket0 = np.array([[1],[0]] ,dtype=complex)
ket1 = np.array([[0],[1]] ,dtype=complex)
ket0,ket1
```

# 3.Superposition States
Quantum superposition  means quantum system can be exist in combination of multiple possible state simulteneously until measuement is made.

Example: $|Ψ> = \frac{1}{\sqrt{2}}(|0> + |1>)$

```python
psi = (1/np.sqrt(2)) * (ket0 + ket1)
psi
```

## 4.Norm of Quantum state 
Quantum state Must be Normalized.so we perfrom norm operation on every quantum state to represent its validity.Quantum state must have **norm = 1**.

```python
norm = np.linalg.norm(psi)
norm
```

## 5. Inner product (<bra|ket>)
inner product is like vector dot product that we previously seen.it tell us how two system  similar.

```python
inner = np.vdot(ket0,ket1)
inner
```

## 6.Tensor Product(⊗)
Tensor product like N X N matrix.it is used to build multi-state qubit system and create another composite system.

```python
two_qubit_state =np.kron(ket0 ,ket1)
two_qubit_state
```

## 7.Matrices for Quantum Gates
Quantum gate are represented in form of matrices.
Example:   
$X = \begin{pmatrix}0 & 1 \\\\ 1 & 0 \end{pmatrix}$
 
$H = \frac{1}{\sqrt{2}}\begin{pmatrix}1 & 1 \\\\ 1 & -1\end{pmatrix}$

```python
X = np.array([[0,1],[1,0]] , dtype=complex)
H = (1/np.sqrt(2))*np.array([[1,1],[1,-1]] , dtype=complex)
X,H
```

## 8.Apply Gate to a Qubit
let apply hadamard gate on Qubit 0 i.e H-Gate on |0> and see what's the result.

```python
H_psi0 = H @ ket0
H_psi0
```

## 9.Check matrix is Unitary
it is mendatory to check matrix is Unitary or not in qunatum computing.Unitary matrix is the square matrix whose conjugate transpose is also its inverse.it help to norm -preserving.
Unitary: $U^{\dagger}U = I$

```python
def is_unitary(U):
        return np.allclose(U.conj().T @ U,np.eye(U.shape[0]))
is_unitary(H),is_unitary(X)
```

## 10.Eigenvalues and Eigenvectors
Quantum gate (unitary matrices) often have eigenvalue +1 or -1 or complex phase .its directly connected to quantum measurement. Lets compute them for Puali-X and Hadamard Gate to see how it operate on it.

```python
eignval_X,eignvect_X=np.linalg.eig(X)
eignval_H,eignvect_H=np.linalg.eig(H)
print("Eigenvalues of X:")
print(eignval_X)
print("Eigenvectors of X:")
print(eignvect_X)

print("Eigenvalues of X:")
print(eignval_H)
print("Eigenvectors of X:")
print(eignvect_H)
```
