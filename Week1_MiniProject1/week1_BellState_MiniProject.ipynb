## Introduction to Quantum State
Qubit can exist in a superposition:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$

Entanglement occures when two qubits cannot be described independently :

$$|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$

We will generate this bell state.

##  1. Import Required libraries

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
```

## 2. Build a 2-Qubit  Quantum Circuit
We will:

1. Apply a Hadamard gate on  qubit 0 $ {\rightarrow} $ which produced superposition.
2. Apply a CNOT-Gate (Control=0,target =1) $ {\rightarrow} $ entangles a Qubit.

This Produce the Bell state $ |\Phi^+\rangle $

```python
# Create 2-Qubit qunatum circuit
qc= QuantumCircuit(2,2)

# step1 : Create superposition on Qubit 0
qc.h(0)

# step2 : Entangle Qubit 0 with Qubit 1
qc.cx(0,1)

#perform Measurement
qc.measure([0,1],[0,1])

#display circuit
qc.draw('mpl')
```

## 3. Simulate and Collect Measurement result
If entanglement work correctly then measurement outcome will be :

- only 00 and 11 should be appear
- No 01 and 10

This shows that measurement of one qubit instantly determined the other, demonstrating the entanglement.

```python
# Simulate on local Statevector

sim = AerSimulator()
result = sim.run(qc,shots= 1024).result()
counts = result.get_counts()
counts
```

## 4. Visualize the Result
visualize this result by plotting histogram .

```python
plot_histogram(counts,figsize=(4,4))
```

## 5.Interpretation of results
our histogram shows two tall bar for '00' and '11' ,This means,
- Qubit are **correlated**
- They collapse together into same value
- This is sinature of Qunatum Entaglement

Because we cannot assign independent state to each qubit ,this proves that state is entangled.
