# Introduction to Quantum computing
This notebook demonstrates a simple quantum program using Qiskit:
1. create a quantum circuit with 1 qubit.  
2. Apply Hadamard gate to create superposition.  
3. measure the qubit.  
4. Run the circuit on a simulator and visualize the result.  

```python
from qiskit_aer import AerSimulator
from qiskit import QuantumCircuit
#%matplotlib inline
from qiskit.visualization import plot_histogram
```

let's Create Quantum Circuit

```python
#Create a quantum circuit with 1 qubit and 1 classical bit
qc =QuantumCircuit(1,1)
#Apply Hadamard gate to put the qubit into superposition
qc.h(0)
#Measure the qubit
qc.measure(0,0)
#visualize the circuit
qc.draw()
```

Run the circuit on Simulator

```python
# Use AerSimulator
simulator=AerSimulator()

# Execute the circuit on the simulator
# Get results
result=simulator.run(qc).result()
counts=result.get_counts()

# print results
print(counts)

# plot histogram
plot_histogram(counts)
```

### Explanation
- Qubit start in |0> state 
- Hadamard gate create superposition : $$ \frac{|0\rangle + |1\rangle}{\sqrt{2}} $$
- Measuring collapse the qubit into 0 and 1.
- Running multiple shot shows roughly equal  0s and 1s.
