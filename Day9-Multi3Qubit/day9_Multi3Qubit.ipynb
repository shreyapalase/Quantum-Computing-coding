# Multi-Qubit System in Quantum Computing Part-2( 3-Qubit Gates)
Previously We studied  2-qubit  gates. now this notebook demonstrate implementation of 3-Qubit system.In this notebook we studied following 3-qubit gates:
- Toffoli gate(CCX)
- controlled-SWAP(fredkin)gate
- CCZ gate
- Multi-Controlled rotations(CRX,CRY,CRZ,CU)
- Custom Unitary 3-Qubit Unitary gate

All simulates using **Qiskit Statevector**

# 1.Import  Required Qiskit Libraries

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector,Operator
from qiskit.visualization import plot_state_qsphere,plot_histogram
from qiskit_aer import AerSimulator

import numpy as np
```

## 2.Helper Function to Print Statevector
let's create state vector function to print statevector state.

```python
def show_state(qc):
    sv=Statevector.from_instruction(qc)
    print('\n',sv)
    return sv
```

## 3. Prepare Basic States
We will use some states for testing gates

```python
# |010> 
qc_010 =QuantumCircuit(3)
qc_010.x(1)
show_state(qc_010)
qc_010.draw('mpl')
```

## 4.Standerd 3-Qubit Gates

### 4.1 Toffoli Gate(CCX)
Flips the state of the target qubit if both controls qubits are |1>.

```python
qc=QuantumCircuit(3)
qc.ccx(0,1,2)
sv_ccx=show_state(qc)
qc.draw('mpl')
```

### 4.2 Controlled-SWAP  (Fredkin Gate)
Swaps the state of two qubits if the control qubit is |1>

```python
qc=QuantumCircuit(3)
qc.cswap(0,1,2)
sv_cswap=show_state(qc)
qc.draw('mpl')
```

### 4.3 CCZ Gate
Applies a Z(Phase flip) to the target qubit if both control qubits are |1>

```python
qc=QuantumCircuit(3)
qc.h(2)
qc.ccz(0,1,2)
qc.h(2)
sv_ccz=show_state(qc)
qc.draw('mpl')
```

### 4.4 Multi-Controlled Rotations(MCRX,MCRY,MCRZ)

### 1. MCRX(Multi-controlled RX gate)
Rotates the target qubit around the X-axis by a given angle if allc ontrol qubit are |1>

```python
qc=QuantumCircuit(3)
qc.mcrx(np.pi/3 ,[0,1],2)
sv_mcrx=show_state(qc)
qc.draw('mpl')
```

### 2. MCRY(Multi-Controlled RY-Gate)
Rotates the target qubit around the Y-axis by a given angle if all control qubit are |1>

```python
qc=QuantumCircuit(3)
qc.mcry(np.pi/3 ,[0,1],2)
sv_mcry=show_state(qc)
qc.draw('mpl')
```

### 3. MCRZ(Multi_Controlled RZ-Gate)
Rotates the target qubit around the Z-axis by a given angle if all control qubit are |1>

```python
qc=QuantumCircuit(3)
qc.mcrz(np.pi/3 ,[0,1],2)
sv_mcrz=show_state(qc)
qc.draw('mpl')
```

### 4. CP(Multi-Controlled Phase Gate)
Applies a phase shift to the target qubit if all qubit are |1>

```python
qc=QuantumCircuit(3)
qc.mcp(np.pi/2 ,[0,1],2)
sv_mcp=show_state(qc)
qc.draw('mpl')
```

## 5. Create Custom 3-Qubit Unitary Gate
User defined unitary acting simulteneously on all three qubit

```python
## 5. Create Custom 3-Qubit Unitary Gate
theta= np.pi/4
U=np.eye(8,dtype=complex)
U[7,7] =np.exp(1j * theta)
qc=QuantumCircuit(3)
qc.unitary(Operator(U),[0,1,2])
qc.draw('mpl')
```

## 6.Statevector Representation of 3-Qubit State
 view the final quantum state of 3-Qubit system.

```python
qc_demo=QuantumCircuit(3)
# create some superposition and Entanglement
qc_demo.h(0)
qc_demo.cx(0,1)
qc_demo.ccx(0,1,2)
state= Statevector.from_instruction(qc_demo)
print(state)
qc_demo.draw('mpl')

```

 you also plot the blochvector and statevector to visualize Qubit

```python
plot_state_qsphere(qc_demo,figsize=(4,4))
```

## 7. Use Simulator for 3-Qubit State

```python
from qiskit_aer import AerSimulator
qc=QuantumCircuit(3,3)
qc.x(0)
qc.x(1)
qc.ccx(0,1,2)
qc.measure([0,1,2],[0,1,2])
backend=AerSimulator()
result=backend.run(qc,shots=1024).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
```

You can see, result of our histrogram for state |111> is 100% .
