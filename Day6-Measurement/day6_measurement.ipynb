# Measurement And Probabilities
Welcome today we learn  one of the most importtant concept in quantum computing:
**Measurement and Probability**

In Quantum computing :
- Before Measurement , A qubit can be in  **superposition**.
- After Measurement , it becomes a **classical value (0 or 1)**
- The result is **Probabilistic** , not deterministic.

This Notenook will guide you step by step.

1. What is measurement?
2. What is Probability in Quantum state?
3. Create superposition using the hadamard gate?
4. Measure the qubit many time?
5. See the probability distribution?
6. Explore Multi-qubit Measurement?

Let's Begin!


## Step 1 - Import Qiskit and Tools
We need only few basic modules.

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
```

## Step 2 - What is Qubit before Measurement
A Qubit can be :
|0>,|1> or any combination a|0> +b|1>

Example of Superposition state:

$$
    |\psi> = 0.707|0> + 0.707|1>
$$

This state means :
- Probability of |0> = 0.5
- Probabbility of |1> = 0.5

Let's create a superposition.


## Step 3 - Create Superposition Using The Hadamard Gate
The **Hadamard(H)**  gate creates equal superposition:
$$
H|0> = \frac{|0> + |1>}{\sqrt{2}}
$$

```python
qc = QuantumCircuit(1,1)
qc.h(0)
qc.measure(0,0)
qc.draw('mpl')
```

## Step 4 - Measure the Qubit Many Times
Because Quantum Measurement is **probabilistic** , we must measure many time to see probability.
We will run the circuit **1000**times.

```python
simulator=AerSimulator()
result=simulator.run(qc).result()
counts=result.get_counts()
print(counts)
```

## Step 5 - Plot Probability Distribution
 The Histogram shows  how many  times the result was 0 or 1.

```python
plot_histogram(counts)
```

You should see roughly :
- 50 %  $ {\rightarrow} $ 0
- 50 %  $ {\rightarrow} $ 1

This is matches the math.

## Step 6 - Measurement  of |1> state
Let's see what happen if the qubit is **not** in superposition.
This circuit prepares  |1> and Measure it. 

```python
qc2 = QuantumCircuit(1,1)
qc2.x(0)
qc2.measure(0,0)
qc2.draw('mpl')
```

```python
result2 =simulator.run(qc2,shots =500).result()
counts2=result2.get_counts()
print(counts2)
```

You should see:
- **100% probability of getting 1**

## Step 7 - Measuring in different Bases(Z-basis vs X-basis)
so far we measure  in the **Z-basis** , which distinguishes between:
- |0>
- |1>

But Quantum computer can meausre in other bases too!
### X -basis
The X-basis states are:
- |+> = (|0> + |1>)$ {\sqrt{2}} $
- |-> = (|0> - |1>)$ {\sqrt{2}} $

To measure in X-basis:
1. Apply **H-gate** (it rotates the basis)
2. Then measure normally (in Z-basis)

Let's do it!

```python
qc3=QuantumCircuit(1,1)
qc3.h(0)  # prepare |+>
qc3.h(0)  # rotate mesurement basis
qc3.measure(0,0)
qc3.draw('mpl')
```

```python
result3 =simulator.run(qc3,shots =1000).result()
counts3=result3.get_counts()
print(counts3)
```

```python
plot_histogram(counts3)
```

Expected Result :
- 100% $ {\rightarrow} $ 0
Why?
Because |+> is the "0"  state in the X-basis!
